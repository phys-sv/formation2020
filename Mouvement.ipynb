### Compléter le code ci-dessous en mettant le nom du fichier.txt à ouvrir et le nom du référentiel :

```python
import matplotlib.pyplot as plt


# Extraction des données du fichier texte de pointage, A NE PAS MODIFIER =========

t,x,y=[],[],[] # définition de 3 listes vides pour les 3 variables t, x et y

# ouverture du fichier texte en lecture r=READ avec la méthode with 
with open(".........txt", "rb") as fichierTexte:       
   # extraction des données du fichier ligne par ligne
    i=1
    for ligne in fichierTexte:          
        if i > 3 : # extraction des données à partir de la 4ième ligne du fichier
            # conversion en nombre des données d d'une ligne                        
            ti,xi,yi=[float(d) for d in ligne.decode('latin1').replace(",",".").split(";")]            
            # .replace  remplace les "," par un point "."
            # .split(";") sépare les grandeurs qui étaient séparées par un ; dans le fichier texte    
            # ajout des données ti, xi et yi dans les listes t, x et y       
            t.append(ti), x.append(xi), y.append(yi)   
        i=i+1
        


# Tracé des points de la trajectoire y=f(x)=======================================                   
# Initialisation de la figure, nom de la fenêtre de visualisation

plt.figure('Trajectoire dans le référentiel .............. ')         # titre de la figure
plt.title('Trajectoire d\'un point de la roue ')                 # titre du graphique
plt.xlabel('x(en m)')                                            # légende de l'axe des abscisses
plt.ylabel('y(en m)')                                            # légende de l'axe des abscisses
plt.axis('equal')                                                # Repère orthonormé 
plt.grid(True)                                                   # ajout d'une grille de fond



# Représentation des points de coordonnées (x(t),y(t)),rouge'r',forme'+',taille 6
plt.plot(x,y,'r+',ms=6) 





plt.savefig('Trajectoire dans le référentiel ............pdf')    # sauve le graphique en fichier pdf
plt.show()                                                        # affiche la figure 



```
