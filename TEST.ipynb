```python



import matplotlib.pyplot as plt
import numpy as np
from math import sqrt

#===============================================================================
# Extraction des données du fichier texte de pointage, A NE PAS MODIFIER 


t,x,y=[],[],[] # définition de 3 listes vides pour les 3 variables t, x et y

# ouverture du fichier texte en lecture r = READ vavec la méthode with 
with open("astro.txt", "r") as fichierTexte:       
   # extraction des données du fichier ligne par ligne
    i=1
    for ligne in fichierTexte:          
        if i > 3 : # extraction des données à partir de la 4ième ligne du fichier
            # conversion en nombre des données d d'une ligne                        
            ti,xi,yi=[float(d) for d in ligne.replace(",",".").split(";")]   
            # .replace  remplace les "," par un point "."
            # .split(";") sépare les grandeurs qui étaient séparées par un ; dans le fichier texte        
            # ajout des données ti, xi et yi dans les listes t, x et y       
            t.append(ti), x.append(xi), y.append(yi)   
        i=i+1
        

                        

#===============================================================================
#Calcul de glune


# Calcul des valeurs de la liste pour Vy    
Vy=[0]                            # Crée une liste qui contient 0 pour le point 0.
for i in range(1,len(t)-1) :      # Pour créer une liste de len(t)-2 éléments en commançant par i=1
                                 
    
    # Calcul de la coordonnée Vyi du vecteur vitesse Vi sur l'axe y au point n°i
    Vyi=(y[i+1]-y[i-1])/(t[i+1]-t[i-1])            
    Vy.append(Vyi) # ajout des nouveaux à la liste : la liste contient en tout len(t)-1 éléments car 0 plus les autres



# Calcul des valeurs de la liste de gy
gy=[0,0]                        # Crée une liste qui contient 0 pour les point 0 et 1
for i in range(2,len(t)-2) :     # Pour créer une liste de len(t)-4 éléments en commançant par i=2                   
   
    gyi=(Vy[i+1]-Vy[i-1])/(t[i+1]-t[i-1])
    gy.append(gyi)              # Ajout des nouveaux à la liste (en tout len(t)-2 éléments dont 0 et 0 qu'il faut enlever)


del(gy[0])
del(gy[0])
   
   
   
# L'axe oy est orienté vers le haut donc g = -gy   (car gy<0)
gy=np.array(gy)   #Transforme la liste en tableau pour pouvoir effectuer des calculs
g=-gy


# Calcul de la moyenne des valeurs de g
gL=sum(g)/len(g)
print ('Valeur non arrondie de l\'intensité de pesanteur sur la lune :',gL,'N/kg')




#===============================================================================
# Histogramme des valeurs de g
plt.hist(g,bins=12,range=(-6,6),facecolor="c", edgecolor = 'black') #on trace l'histogramme des valeurs de l'indice 
plt.title('histogramme des valeurs de glune-Fichier2')
plt.xlabel('glune (N/kg)')
plt.ylabel('Nombre de mesures')
plt.savefig('Histogramme des valeurs de glune-Fichier2.pdf')

#incertitudes

sigma=sqrt((1/(len(g)-1))*sum((g-gL)**2))
U=sigma/sqrt(len(g))

  
print ('Valeur non arrondie de l\'écart type :',sigma,'N/kg')
print ('Valeur non arrondie de U(g) :',U,'N/kg')

plt.show()

```
